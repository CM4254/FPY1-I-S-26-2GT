```python
print("VAMOS CALCULAR SEU IMC?")
ALTURA = int(input("Sua Altura "))
PESO = int(input("Seu Peso "))
imc = round((PESO/((ALTURA*2)/100)),2)
if (imc  < 18.5):
  print(f"Seu imc é {imc} - Abaixo do peso")
elif (imc >= 18.5 and imc < 25):
   print(f"Seu imc é {imc} -  Peso normal")
elif ( imc >= 25 and imc <30):
   print(f"Seu imc é {imc} - Sobrepeso")
elif ( imc >= 30 and imc <40):
   print(f"Seu imc é {imc} - Obesidade")
else:
     print(f"Seu imc é {imc} - Obesidade Grave")
```

```python
from datetime import date
ANO_ATUAL  = int(date.today().year)
MES_ATUAL = int(date.today().month)
DIA_ATUAL = int(date.today().day)

print("Digite o ano do seu nascimento: ")
ano = int(input())
print("Digite o mês do seu nascimento: ")
mes = int(input())
print("Digite o dia do seu nascimento: ")
dia = int(input())


IDADEDIF =  (ANO_ATUAL - ano) -1
IDADE =  (ANO_ATUAL - ano)

if (MES_ATUAL < mes) :
    print(f"sua idade é {IDADEDIF} anos")
elif (MES_ATUAL == mes) and (DIA_ATUAL < dia):
    print(f"sua idade é {IDADEDIF} anos")
else:
    print(f"sua idade é {IDADE} anos")

```

```python
palavra = input("Digite sua palavra ")
vogais = {"A", "E", "I", "O", "U", "a", "e", "i", "o", "u"}
consoantes = {
    "B","C","D","F","G","H","J","K","L","M","N","P","Q","R","S","T","V","W","X","Y","Z",
    "b","c","d","f","g","h","j","k","l","m","n","p","q","r","s","t","v","w","x","y","z"
}

vogaisc = 0
consoantesc = 0

for caract in palavra:

    if caract in vogais:

      vogaisc +=1

for caract in palavra:

    if caract in consoantes:

      consoantesc +=1

print(f"A quantidade de vogais é {vogaisc}")
print(f"A quantidade de consoantes é {consoantesc}")

```
